```python
import pandas as pd
import matplotlib.pyplot as plt

# Örnek veri oluşturma (eğer elinizde veri yoksa)
data = pd.DataFrame({
    'prcp': [0, 0.1, 0.5, 1.0, 5.0, 10.0],  # Yağış miktarı (mm)
    'steps': [1000, 2000, 3000, 4000, 5000, 6000]  # Günlük adım sayısı
})

# Yağış ve adım sayısını scatter plot ile göster
plt.scatter(data['prcp'], data['steps'], alpha=0.6)
plt.title('Yağış Miktarı vs Günlük Adım Sayısı')
plt.xlabel('Yağış Miktarı (mm)')
plt.ylabel('Adım Sayısı')
plt.grid(True)
plt.show()
```
