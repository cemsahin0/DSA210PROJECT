```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Örnek veri oluşturma (daha fazla veri noktasıyla)
np.random.seed(42)
data = pd.DataFrame({
    'tavg': np.random.uniform(0, 30, 100),  # Ortalama sıcaklık (°C)
    'prcp': np.random.uniform(0, 50, 100),  # Yağış miktarı (mm)
    'steps': np.random.uniform(5000, 15000, 100)  # Adım sayısı
})

# 3D Grafik
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# X, Y, Z değerleri
sc = ax.scatter(data['tavg'], data['prcp'], data['steps'], c=data['steps'], cmap='viridis', s=50)

# Renk skalası ekleme
plt.colorbar(sc, ax=ax, label='Adım Sayısı')

# Başlık ve etiketler
ax.set_title('Sıcaklık, Yağış ve Adım Sayısı')
ax.set_xlabel('Ortalama Sıcaklık (°C)')
ax.set_ylabel('Yağış Miktarı (mm)')
ax.set_zlabel('Adım Sayısı')

plt.show()
```
